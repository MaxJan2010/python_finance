```python
import MetaTrader5 as mt5
import pandas as pd
import plotly.express as px
import numpy as np
from IPython.display import display
from datetime import datetime
```

```python
# connect to MetaTrader5 as mt5
mt5.initialize()
```

```python
from_date=datetime(2020,1,1)
to_date=datetime.now()
trade_deals=mt5.history_deals_get(from_date, to_date)
display(trade_deals)
```

```python
# display these positions as a table using pandas.DataFrame
deals = pd.DataFrame(list(trade_deals), columns=trade_deals[0]._asdict().keys())

#  Get Columns Titles to check Existing Columns
titles = deals.columns

for title in titles:
 print(title)
```

```python
# re-factor data frame
# convert time in history to human readable time format
deals['time'] = pd.to_datetime(deals['time'], unit='s')

# drop un-usable columns
deals.drop(['ticket', 'order', 'position_id', 'time_msc', 'entry', 'comment', 'external_id'], axis=1, inplace=True)
display(deals)
```

```python
# re-factor data frame
# in column 'type' => replace 
deals['type'] = deals['type'].replace([0], 'buy')
deals['type'] = deals['type'].replace([1], 'sell')
deals['type'] = deals['type'].replace([2], 'balance')

display(deals)
```

```python
# re-factor data frame
# drop orders rows by filter 'type' & 'magic' columns
deals = deals.query("profit != 0")

display(deals)
```

```python
# Find Initial Deposit = Initial Balance 
# deposit = starting_balance  # Use in backtest Factors
deposit = deals.iloc[0]['profit']
print('deposit = ', deposit)
```

```python
deals['pnl'] = deals['profit'].cumsum()
```

```python
# drop balance row in data frame
deals.drop(index=[0], inplace =True)
display(deals)
```

```python
def create_profit_type(profit):
    # Finding Signal
    if profit > 0:
        return 'win'
    elif profit < 0:
        return 'loss'
    
deals['profit_type'] = np.vectorize(create_profit_type)(deals['profit'])

deals.dropna(inplace=True)

display(deals)
```

```python
# find balance
balance = round(deals['pnl'].iloc[-1], 2)
print('balance = ', balance)

# find total net profit
total_net_profit = round(deals['profit'].sum(), 2)  # Use in backtest Factors
print('total net profit = ', total_net_profit)

# find gross profit
gross_profit = round(sum(deals[deals['profit_type']=='win']['profit']), 2)
print('gross profit = ', gross_profit)

# find gross loss
gross_loss = round(sum(deals[deals['profit_type']=='loss']['profit']), 2)
print('gross loss = ', gross_loss)

# find profit factor
profit_factor = abs(round((gross_profit / gross_loss), 2))
print('profit factor = ', profit_factor)

# find largest profit trade
largest_profit_trade = round(deals['profit'].max(), 2)
print('largest profit trade = ', largest_profit_trade)

# find largest loss trade
largest_loss_trade = round(deals['profit'].min(), 2)
print('largest loss trade = ', largest_loss_trade)

# find total trades
total_trades = len(deals)
print('total trades = ', total_trades)

# find total count of SELL trades
count_sell_trades = len(deals)
print('count SELL trades = ', count_sell_trades)
```
