```python
# import required libraries

import MetaTrader5 as mt5
import pandas as pd
import plotly.express as px
import numpy as np
import datetime
from IPython.display import display
from datetime import datetime
```

```python
# connect to MetaTrader5 as mt5
mt5.initialize()
```

```python
from_date=datetime(2020,1,1)
to_date=datetime.now()
trade_deals=mt5.history_deals_get(from_date, to_date)
display(trade_deals)
```

```python
# display these positions as a table using pandas.DataFrame
deals = pd.DataFrame(list(trade_deals), columns=trade_deals[0]._asdict().keys())

deals
```

```python
# re-factor data frame
# convert time in history to human readable time format
deals['time'] = pd.to_datetime(deals['time'], unit='s')

deals
```

```python
#  Get Columns Titles to check Existing Columns
titles = deals.columns

for title in titles:
 print(title)
```

```python
# drop un-usable columns
deals.drop(['ticket', 
            'order',  
            'time_msc', 
            'entry', 
            'comment', 
            'external_id'
            ], 
           axis=1, inplace=True)

deals
```

```python
# collect positions open status

# copy dataframe (deals) to new dataframe (deals_open)
deals_open = deals.copy()

# separate positions open status in separated dataframe 
deals_open = deals_open.drop_duplicates(subset=["position_id"], keep="first")

#  drop un-usable columns in dataframe (deals_open)
deals_open.drop(['type', 
                'commission',  
                'swap', 
                'profit', 
                'fee',
                ], 
               axis=1, inplace=True)

# add "open_" to each column name
# deals_open = deals_open.add_prefix('open_')

# add "open_" to each column name except 'position_id'
# deals_open.columns = deals_open.columns.map(lambda x : x+'_open' if x !='position_id' else x)

# add "open_" to each specific columns name
deals_open.columns = ['open_' + col if col in ['time', 'price'] else col for col in deals_open.columns]

deals_open
```

```python
# collect positions close status

# copy dataframe (deals) to new dataframe (deals_close)
deals_close = deals.copy()

# separate positions close status in separated dataframe 
deals_close = deals_close.drop_duplicates(subset=["position_id"], keep="last")

#  drop un-usable columns in dataframe (deals_open)
deals_close.drop(['magic',
                'reason',
                'volume',
                'symbol',
                ], 
               axis=1, inplace=True)

# add "close_" to each column name
# deals_close = deals_close.add_prefix('close_')

# add "close_" to each column name except 'position_id'
# deals_close.columns = deals_close.columns.map(lambda x : x+'_close' if x !='position_id' else x)

# add "open_" to each specific columns name
deals_close.columns = ['close_' + col if col in ['time', 'price'] else col for col in deals_close.columns]

deals_close
```

```python
# merge (open & close) data frames into new dataframe (positions)

positions = deals_open.merge(deals_close, how='inner', on='position_id')

positions
```

```python
#  Get Columns Titles to check Existing Columns
titles = positions.columns

for title in titles:
 print(title)
```

```python
# export dataframe to csv file

# create file name based on current date and time
filename = datetime.now().strftime('%Y-%m-%d-%H-%M-%S.csv')

# export dataframe to csv file
# positions.to_csv(filename)
```

```python
# re-factor dataframe
# in column 'type' => replace 
positions['type'] = positions['type'].replace([0], 'sell')
positions['type'] = positions['type'].replace([1], 'buy')
positions['type'] = positions['type'].replace([2], 'balance')

positions
```

```python
# Find Initial Deposit = Initial Balance 
# deposit = starting_balance  # Use in backtest Factors
deposit = positions.iloc[0]['profit']
print('deposit = ', deposit)
```

```python
positions['pnl'] = positions['profit'].cumsum()

# positions.eval('total_pnl = commission + fee + pnl')
```

```python
# create a column with timedelta (duration_days) including days
# positions['duration_days'] = (positions.close_time - positions.open_time)

# create a column with timedelta (duration_hours) as total hours, as a float type
# positions['duration_hours'] = (positions.close_time - positions.open_time) / pd.Timedelta(hours=1)

# create a column with timedelta as total duration_minutes, as a float type
# positions['duration_minutes'] = (positions.close_time - positions.open_time) / pd.Timedelta(minutes=1)

# create a column with timedelta as total duration_seconds, as a float type
positions['duration_seconds'] = (positions.close_time - positions.open_time) / pd.Timedelta(seconds=1)

positions
```

```python
# drop balance row in data frame
# positions = positions.drop(0)
positions.drop(index=[0], inplace =True)
display(positions)
```

```python
# create profit type column (win / loss)

def apply_profit_type(x):
# Finding profit type
  if x['profit'] >= 0: 
      return "win"
  elif x['profit'] < 0:
      return "loss"
  else: return None

positions['profit_type'] = positions.apply(apply_profit_type, axis=1)

# positions.dropna(inplace=True)

display(positions)
```

```python
# find balance
balance = round(positions['pnl'].iloc[-1], 2)
print('balance = ', balance)

# find total net profit
total_net_profit = round(positions['profit'].sum(), 2)  # Use in backtest Factors
print('total net profit = ', total_net_profit)

# find gross profit
gross_profit = round(sum(positions[positions['profit_type']=='win']['profit']), 2)
print('gross profit = ', gross_profit)

# find gross loss
gross_loss = round(sum(positions[positions['profit_type']=='loss']['profit']), 2)
print('gross loss = ', gross_loss)

# find profit factor
profit_factor = abs(round((gross_profit / gross_loss), 2))
print('profit factor = ', profit_factor)

# find largest profit trade
largest_profit_trade = round(positions['profit'].max(), 2)
print('largest profit trade = ', largest_profit_trade)

# find largest loss trade
largest_loss_trade = round(positions['profit'].min(), 2)
print('largest loss trade = ', largest_loss_trade)

# find total trades
total_trades = len(positions)
print('total trades = ', total_trades)

# find total count of SELL trades
count_sell_trades = positions['type'].value_counts()['sell']
print('count SELL trades = ', count_sell_trades)

# find total count of BUY trades
count_buy_trades = positions['type'].value_counts()['buy']
print('count BUY trades = ', count_buy_trades)

# find total count of WIN trades
count_win_trades = positions['profit_type'].value_counts()['win']
print('count WIN trades = ', count_win_trades)

# find total count of LOSS trades
count_loss_trades = positions['profit_type'].value_counts()['loss']
print('count LOSS trades = ', count_loss_trades)

# select rows by value
# avg_win = deals_group.loc[deals_group['profit_type'] == 'win']
# print('avg_win' , avg_win)

# find average value of WIN Trades
ave_win_value = round(positions.loc[positions['profit_type'] == 'win']['profit'].mean() , 2)
print('ave_win_value = ' , ave_win_value)

# find average value of LOSS Trades
ave_loss_value = round(positions.loc[positions['profit_type'] == 'loss']['profit'].mean(), 2)
print('ave_loss_value = ' , ave_loss_value)
```

```python
# Collect consecutive Profit / Loss (Factors)

# find consecutive Profit / Loss
positions_group = positions.groupby((positions.profit_type != positions.profit_type.shift()).cumsum()).agg(
  profit_type = ('profit_type', 'first'),
  grand_profit = ('profit', 'sum'),
  count = ('profit_type', 'count')
).reset_index(drop=True)

print('count trades groups = ')
print(positions_group)

# find largest consecutive WIN trade
largest_cons_win = round(positions_group['grand_profit'].max(), 2)
print('largest consecutive WIN trade = ', largest_cons_win)

# find largest consecutive WIN trade (count)
count_largest_win = positions_group.loc[positions_group['grand_profit'].idxmax(), 'count']
print('largest consecutive WIN trade (count) = ', count_largest_win)

# find largest consecutive LOSS trade
largest_cons_loss = round(positions_group['grand_profit'].min(), 2)
print('largest consecutive LOSS trade = ', largest_cons_loss)

# find largest consecutive LOSS trade (count)
count_largest_loss = positions_group.loc[positions_group['grand_profit'].idxmin(), 'count']
print('largest consecutive LOSS trade (count) = ', count_largest_loss)
```
